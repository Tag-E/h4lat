# Operators free of mixing with lower or equal dimensional operators

This notebook displays the operator selections returned by `get_op_selection(n_der)`,
which provides H(4)-irreducible operators which have been used in previous works.

Each operator is shown with H(4) irrep, C-parity.

```python
from IPython.display import display, Markdown
from h4lat import get_op_selection
```

---
## 1 derivative (see [2401.05360](https://arxiv.org/abs/2401.05360))

Operators with 1 derivative: 2 vector (irrep (3,1) and (6,3)), 2 axial (irrep (6,4)),
and 4 tensor (irreps (8,1) and (8,2)).

```python
ops1 = get_op_selection(n_der=1)
```

### Vector — irreps (3,1) and (6,3)

```python
for op in [op for op in ops1 if op.X == 'V']:
    display(Markdown(f"**irrep: {op.irrep} &emsp; C-parity: {op.C} &emsp; Trace condition: {op.tr}**"))
    op.display()
    display(Markdown("&nbsp;"))
```

### Axial — irrep (6,4)

```python
for op in [op for op in ops1 if op.X == 'A']:
    display(Markdown(f"**irrep: {op.irrep} &emsp; C-parity: {op.C} &emsp; Trace condition: {op.tr}**"))
    op.display()
    display(Markdown("&nbsp;"))
```

### Tensor — irreps (8,1) and (8,2)

```python
for op in [op for op in ops1 if op.X == 'T']:
    display(Markdown(f"**irrep: {op.irrep} &emsp; C-parity: {op.C} &emsp; Trace condition: {op.tr}**"))
    op.display()
    display(Markdown("&nbsp;"))
```
