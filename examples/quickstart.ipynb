## 1. H(4) irreps at a glance

H(4) has order 384 and **20 irreducible representations** labelled `(k, l)` where `k`
is the dimension and `l` distinguishes inequivalent irreps of the same dimension.
The full list is exposed as `rep_label_list`.

````python
from h4lat import rep_label_list, irrep_dim, irrep_texname

print(f"Number of irreps: {len(rep_label_list)}")
print()
print(f"{'Label':<8}  {'Dim':>4}  LaTeX name")
print("-" * 38)
for label in rep_label_list:
    print(f"{str(label):<8}  {irrep_dim[label]:>4}  ${irrep_texname[label]}$")
````

## 2. Tensor-product decomposition

`get_multiplicities` decomposes a tensor product of H(4) irreps into a direct sum
using the character-table formula.  It returns a length-20 list of multiplicities.

`latex_print_multiplicities` formats the same result as a LaTeX equation and,
when running inside a notebook, renders it inline with `display_eq=True`.

````python
from h4lat import get_multiplicities, latex_print_multiplicities, rep_label_list

# Decompose (4,1) ⊗ (4,1)  — vector ⊗ vector
muls = get_multiplicities((4, 1), (4, 1))

print("Non-zero multiplicities in (4,1) ⊗ (4,1):")
for i, m in enumerate(muls):
    if m > 0:
        print(f"  {rep_label_list[i]}  ×{m}")
````

````python
# Render the decomposition as a LaTeX equation
latex_string = latex_print_multiplicities((4, 1), (4, 1), display_eq=True, return_str=True)
print("Latex string for (4,1) ⊗ (4,1):")
print(latex_string)
````

````python
# Decompose (6,1) ⊗ (4,1) ⊗ (4,1)  — vector ⊗ vector
muls = get_multiplicities((6, 1), (4, 1), (4, 1))

print("Non-zero multiplicities in (6,1) ⊗ (4,1) ⊗ (4,1):")
for i, m in enumerate(muls):
    if m > 0:
        print(f"  {rep_label_list[i]}  ×{m}")
````

````python
# Three-fold product — tensor operator with two derivatives
latex_print_multiplicities((6, 1), (4, 1), (4, 1), display_eq=True, return_str=False)
````

## 3. Loading CG coefficients from the bundled database

`cg_calc` is the main entry point.  When the requested tensor product is already
in the bundled database it loads instantly; otherwise it computes and caches the result.

The result object exposes:
* `cg.mul_list` — the multiplicity list (same as `get_multiplicities`).
* `cg.cg_dict` — a dict mapping irrep index → list of CG matrices (one per multiplicity copy).

````python
from h4lat import cg_calc, rep_label_list

# Load from the bundled database — no computation needed
cg = cg_calc((4, 1), (4, 1))

print("CG coefficient matrices for (4,1) ⊗ (4,1):")
print()
for irep_idx, matrices in cg.cg_dict.items():
    label = rep_label_list[irep_idx]
    print(f"  Irrep {label}:  {len(matrices)} matrix/matrices, shape {matrices[0].shape}")
````

````python
# Pretty-print one CG matrix inside the notebook
from h4lat import print_CGmat

# First irrep in the decomposition, first multiplicity copy
first_irep = list(cg.cg_dict.keys())[0]
print(f"CG matrix for irrep {rep_label_list[first_irep]}:")
print_CGmat(cg.cg_dict[first_irep][0])
````

## 4. Computing new CG coefficients

Pass any combination of irrep labels; if the result is not in the database it will
be computed via the Sakata (1974) projection formula and cached automatically.
The cell below uses a product that *is* bundled so it runs instantly.

````python
# (6,1) ⊗ (4,1) ⊗ (4,1)  — tensor operator with two derivatives
cg3 = cg_calc((6, 1), (4, 1), (4, 1))

print("Non-zero irreps in (6,1) ⊗ (4,1) ⊗ (4,1):")
for irep_idx, matrices in cg3.cg_dict.items():
    label = rep_label_list[irep_idx]
    print(f"  {label}  — {len(matrices)} CG matrix/matrices, shape {matrices[0].shape}")
````

## 5. Constructing a lattice operator

The `Operator` class assembles a lattice operator from a CG coefficient tensor.
It computes and exposes:
* `op.latex_K` — the tree-level kinematic factor K as a LaTeX string.
* `op.C` — C-parity (+1 or -1).
* `op.tr` — trace condition (whether the trace vanishes).
* `op.symm` — index-permutation symmetry.
* `op.evaluate_K(...)` — numerical evaluation of K at given kinematics.

`cg_remapping` reshapes a CG column vector into the rank-n tensor that `Operator` expects.

````python
import numpy as np
from h4lat import cg_calc, cg_remapping, Operator, rep_label_list

cg = cg_calc((4, 1), (4, 1))

# Take the first CG matrix for the first irrep in the decomposition
irep_idx = list(cg.cg_dict.keys())[0]
block    = cg.cg_dict[irep_idx][0]          # shape (16, k)
col      = np.round(block[:, 0], decimals=15)

cgmat = cg_remapping(col, n=2)               # reshape to (4, 4)

op = Operator(
    cgmat       = cgmat,
    id          = 1,
    X           = 'V',
    irrep       = rep_label_list[irep_idx],
    block       = 1,
    index_block = 1,
)

print(op)                         # LaTeX expression for the operator
print()
print("K  =", op.latex_K)         # Kinematic factor
print("C  =", op.C)               # C-parity
print("tr =", op.tr)              # Trace condition
print("symm =", op.symm)          # Index symmetry

#display operator
print("\n----------------------------------")
print("Operator form:")
op.display()
print("Kinematic factor:")
op.K
````

## 6. Loading operators from the bundled database

The package ships with a pre-built operator database covering V, A, and T operators
up to three indices.  `get_OperatorList` and `get_OperatorDict` load it without
any path configuration.

### 6a. Flat list — `get_OperatorList`

````python
from h4lat import get_OperatorList

#read the full list of operators from the bundled database
ops = get_OperatorList()
print(f"Total operators in the bundled database: {len(ops)}")
print()
````

````python
# Inspect the first operator
op = ops[0]
print(f"id={op.id}  X={op.X}  irrep={op.irrep}  block={op.block}")
print(f"Kinematic factor K = {op.latex_K}")
print(f"C-parity  = {op.C}")
print(f"Trace     = {op.tr}")
print(f"Symmetry  = {op.symm}")

#display operator
print("\n----------------------------------")
print("Operator form:")
op.display()
print("Kinematic factor:")
op.K
````

````python
# Numerical evaluation of K at specific kinematics
K_val = op.evaluate_K(m_value=0.5, E_value=1.0, p1_value=0.0, p2_value=0.0, p3_value=0.3)
print(f"K (numerical) = {K_val}")
````

````python
# Quick summary of all operators
print(f"{'id':>4}  {'X':>2}  {'irrep':<8}  {'block':>5}  {'symm':<20}  C")
print("-" * 55)
for o in ops:
    print(f"{o.id:>4}  {o.X:>2}  {str(o.irrep):<8}  {o.block:>5}  {o.symm:<20}  {o.C}")
````

### 6b. Nested dict — `get_OperatorDict`

The dict is keyed first by `(n, X)` (number of indices and Dirac structure),
then by `(irrep, block)`.

````python
from h4lat import get_OperatorDict

d = get_OperatorDict()

print("Available (n, X) keys:", sorted(d.keys()))
````

````python
# All 2-index vector operators in irrep (6,1), multiplicity block 1
# (4,1)⊗(4,1) decomposes into (1,1)⊕(3,1)⊕(6,1)⊕(6,3); (6,1) is the symmetric traceless piece
ops_V2 = d[(2, 'V')][(6, 1), 1]
print(f"2-index Vector operators in (6,1) block 1: {len(ops_V2)} operator(s)")
for o in ops_V2:
    print(f"  id={o.id}  symm={o.symm}  C={o.C}")
````

````python
# All 3-index axial operators in irrep (8,1), block 1
if (3, 'A') in d and ((8, 1), 1) in d[(3, 'A')]:
    ops_A3 = d[(3, 'A')][(8, 1), 1]
    print(f"Number of A3 operators in (8,1) block 1: {len(ops_A3)}")
    for o in ops_A3:
        print(f"  id={o.id}  symm={o.symm}  C={o.C}")
else:
    print("(8,1) block 1 not present for (3, 'A') in the bundled database.")
````
