# Operator catalogue

A complete listing of every operator in the bundled database, organised by
Dirac structure and number of indices, then by H(4) irrep.

The database is loaded as a nested dict via `get_OperatorDict()`:
```
d[(n, X)][(irrep, block)]  →  list[Operator]
```
where `n` is the number of Lorentz indices, `X` is the Dirac structure
(`'V'`, `'A'`, or `'T'`), `irrep` is the H(4) irrep label `(k, l)`,
and `block` is the multiplicity index.

For each operator we display:
* **id** — integer identifier in the bundled database
* **Operator form** — $\sum_\mu c_\mu\, O^X_\mu$ rendered as LaTeX via `op.display()`
* **Kinematic factor** — tree-level K via `op.K`

````python
from IPython.display import display, Markdown
from h4lat import get_OperatorDict

d = get_OperatorDict()
_X_NAMES = {'V': 'Vector', 'A': 'Axial', 'T': 'Tensor'}

total = sum(len(ops) for irreps in d.values() for ops in irreps.values())
print(f"Database loaded — {total} operators across {len(d)} (n, X) groups.")
````

---
## Vector operators — 2 indices &nbsp;&nbsp;`(n=2, X='V')`

````python
for (irrep, block) in sorted(d[(2, 'V')].keys()):
    ops = d[(2, 'V')][(irrep, block)]
    display(Markdown(f"### Irrep **{irrep}**, block {block} &nbsp; ({len(ops)} operator{'s' if len(ops) > 1 else ''})"))
    for op in ops:
        display(Markdown(f"**id = {op.id}** &emsp; C-parity: {op.C} &emsp; Symmetry: {op.symm}  &emsp; Trace condition: {op.tr}"))
        print("Operator form:")
        op.display()
        print("Kinematic factor:")
        display(op.K)
        display(Markdown("&nbsp;"))
````
