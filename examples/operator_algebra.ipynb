# Operator algebra

Operators in the same H(4) irrep form a **vector space over the reals**: you can add
them, subtract them, and scale them by numeric coefficients.  The result is always
another `Operator` object with a consistently updated CG tensor, operator expression,
and kinematic factor.

This notebook demonstrates every algebraic operation supported by the `Operator` class,
showing the operator form and kinematic factor before and after each operation.

Supported operations:
| Expression | Meaning |
|------------|---------|
| `a + b`    | Sum of two operators (same X and index count) |
| `a - b`    | Difference |
| `c * a`    | Scale by real number `c` |
| `a / c`    | Divide by real number `c` |
| `-a`       | Negation |

---
## Setup — load two operators from the same irrep

We pick two 2-index vector operators that both transform under the irrep **(6,1)**,
multiplicity block 1.  Because they share the same irrep and Dirac structure, their
linear combinations are well-defined operators.

```python
from h4lat import get_OperatorDict

d = get_OperatorDict()

# 2-index Vector operators in irrep (3,1), multiplicity block 1
ops = d[(2, 'V')][(3, 1), 1]

op_a = ops[0]   # id=5
op_b = ops[1]   # id=6

print(f"op_a: id={op_a.id}  irrep={op_a.irrep}  block={op_a.block}  symm={op_a.symm}  C={op_a.C}")
print(f"op_b: id={op_b.id}  irrep={op_b.irrep}  block={op_b.block}  symm={op_b.symm}  C={op_b.C}")
```

---
## 1. Inspecting a single operator

The two key display methods:
* `op.display()` — renders the full operator sum $\sum c_{\mu\nu}\, O^X_{\mu\nu}$ as LaTeX.
* `op.K` — the tree-level kinematic factor as a SymPy expression (renders as LaTeX in-notebook).

```python
print("Operator form:")
op_a.display()
```

```python
print("Kinematic factor:")
op_a.K
```

---
## 2. Scalar multiplication: `c * op`

Multiplying by a real coefficient scales all CG coefficients uniformly.
The kinematic factor is *unchanged* — it depends only on the Dirac structure
and kinematics, not on the normalisation of the CG tensor.

```python
op_scaled = 2 * op_a

print("Before (op_a):")
op_a.display()
print("Kinematic factor:")
display(op_a.K)
```

```python
print("After (2 * op_a):")
op_scaled.display()
print("Kinematic factor:")
display(op_scaled.K)
```

Division works symmetrically:

```python
op_half = op_a / 2

print("op_a / 2:")
op_half.display()
print("Kinematic factor:")
display(op_half.K)
```

---
## 3. Negation: `-op`

```python
op_neg = -op_a

print("Before (op_a):")
op_a.display()
print("Kinematic factor:")
display(op_a.K)
```

```python
print("After (-op_a):")
op_neg.display()
print("Kinematic factor:")
display(op_neg.K)
```

---
## 4. Addition: `op_a + op_b`

Both operators must share the same Dirac structure `X` and number of indices.
When they also share the same irrep and block, the result inherits those labels;
otherwise `irrep` and `block` are set to `None` to signal a mixed-irrep combination.

```python
print("op_a:")
op_a.display()
print("Kinematic factor:")
display(op_a.K)
```

```python
print("op_b:")
op_b.display()
print("Kinematic factor:")
display(op_b.K)
```

```python
op_sum = op_a + op_b

print(f"op_a + op_b  →  irrep={op_sum.irrep}  block={op_sum.block}")
print()
print("Operator form:")
op_sum.display()
print("Kinematic factor:")
display(op_sum.K)
```

---
## 5. Subtraction: `op_a - op_b`

```python
op_diff = op_a - op_b

print(f"op_a - op_b  →  irrep={op_diff.irrep}  block={op_diff.block}")
print()
print("Operator form:")
op_diff.display()
print("Kinematic factor:")
display(op_diff.K)
```

---
## 6. Linear combination

All operations can be chained.  Here we build the combination
$3\,O_a - \tfrac{1}{2}\,O_b$ from the same irrep.

```python
op_lc = 2 * op_a - 0.5 * op_b

print(f"3 * op_a - 0.5 * op_b  →  irrep={op_lc.irrep}  block={op_lc.block}")
print()
print("Operator form:")
op_lc.display()
print("Kinematic factor:")
display(op_lc.K)
```

---
## 7. Mixing multiple operators from the same group

We can combine all operators in the group at once using a comprehension.

```python
import functools, operator as pyop

# Equal-weight sum of all operators in (6,1) block 1
op_total = functools.reduce(pyop.add, ops)

print(f"Sum of all {len(ops)} operators in (6,1) block 1:")
print()
print("Operator form:")
op_total.display()
print("Kinematic factor:")
display(op_total.K)
```

---
## 8. Mixing operators from different irreps
Adding operators from *different* irreps is algebraically permitted — the sum is a
valid `Operator` object but its `irrep` is set to `None` to indicate the result does
not transform irreducibly under H(4).

A `ValueError` is raised only if the operators have a different Dirac structure `X`
or a different number of indices `n`.

```python
ops_other_irrep = d[(2, 'V')][(6, 1), 1]   # different irrep: (6,1), same X='V', same n=2
op_other = ops_other_irrep[0]

#adding operators from different irreps is permitted
op_mixed = op_a + op_other
op_mixed.display()
print(f"irrep: {op_mixed.irrep}")

#adding operators with different X or number of indices is not permitted
ops_other_X = d[(2, 'A')][(6, 4), 1]   # different X: (6,4), X='A', same n=2
op_other = ops_other_X[0]

try:
    op_mixed = op_a + op_other
    op_mixed.display()
except ValueError as e:
    print(f"Different X or number of indices → ValueError: {e}")
```
